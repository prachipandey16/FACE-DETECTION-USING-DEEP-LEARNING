```python
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from mtcnn.mtcnn import MTCNN
```

```python
def draw_image_with_boxes(filename,result_list):
    data = pyplot.imread(filename)
    pyplot.imshow(data)
    ax = pyplot.gca()
    for i in range(len(result_list)):
        #print(result_list[i])
        for result in result_list:
            x,y,width,height = result['box']
            rect = Rectangle((x,y),width,height,fill=False,color='cyan')
            ax.add_patch(rect)
        pyplot.show()
```

```python
def draw_image_with_circles(filesname, result_list):
    data = pyplot.imread(filename)
    pyplot.imshow(data)
    ax = pyplot.gca()
    for i in range(len(result_list)):
        for result in result_list:
            x,y,width,height = result['box']
            rect = Rectangle((x,y),width,height,fill=False,color='cyan',linewidth=3)
            ax.add_patch(rect)
            for key,value in result['keypoints'].items():
                dot = Circle(value, radius=6, color='cyan')
                ax.add_patch(dot)
        pyplot.show()
```

```python
def draw_faces(filename, result_list):
    data = pyplot.imread(filename)
    for i in range(len(result_list)):
        x1,y1,width,height = result_list[i]['box']
        x2,y2 = x1+ width, y1+height
        
        pyplot.subplot(1, len(result_list), i+1)
        pyplot.axis('off')
        pyplot.imshow(data[y1:y2, x1:x2])
        # show the plot
    pyplot.show()
```
